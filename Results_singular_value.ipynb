```python
import numpy as np
import pandas as pd
import scipy.stats
import os
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib import rcParams

config = {
            "font.family": 'serif',
            "font.size": 12,# 相当于小四大小
            "mathtext.fontset": 'stix',#matplotlib渲染数学字体时使用的字体，和Times New Roman差别不大
            "font.serif": ['Arial'],#['Times New Roman'],#宋体
            'axes.unicode_minus': False # 处理负号，即-号
         }
rcParams.update(config)
```

```python
td = np.load('./Step2_Koopman/databygroup.npy',allow_pickle=True).tolist()
Alls={}
for f in td.keys():
    tem = np.load('./Step2_Koopman/results/'+f+'/s.npy',allow_pickle=True)
    Alls[f] = tem

AllK={}
for f in td.keys():
    tem = np.load('./Step2_Koopman/results/'+f+'/K.npy',allow_pickle=True)
    AllK[f] = tem
```

# 正文图

```python
colormap = plt.get_cmap('Blues')
colors   = colormap(np.linspace(1, 0.5, Alls['historydata'].shape[0])) # 按百分比

ylabels = ['Historical data','sub1','sub2', 'sub3','sub4','sub5','sub6','sub7', 'sub8','sub9','sub10']
ks = ['historydata','sub1','sub2', 'sub3','sub4','sub5','sub6','sub7', 'sub8','sub9','sub10']
alphas = [0.6 for _ in range(len(ks))]
colormap = plt.get_cmap('Blues')
colors   = colormap(np.linspace(1, 0.1, len(ks))) # 按百分比

fontsizes = [18, 15, 20]

fig = plt.figure(dpi=100,
                    constrained_layout=True,#类似于tight_layout，使得各子图之间的距离自动调整【类似excel中行宽根据内容自适应】
                    figsize=(8,5)
                    )
gs = GridSpec(1, 1, figure=fig)#GridSpec将fiure分为3行3列，每行三个axes，gs为一个matplotlib.gridspec.GridSpec对象，可灵活的切片figure
ax = fig.add_subplot(gs[0, 0])
line = 0
for line in range(len(ks)):
    #ax = fig.add_subplot(gs[line, 0])
    ax.scatter([i for i in range(Alls[ks[line]].shape[0])],Alls[ks[line]],alpha=alphas[line]+0.4,color = colors[line],label=ks[line])
    ax.plot([i for i in range(Alls[ks[line]].shape[0])],Alls[ks[line]],alpha=alphas[line],color = colors[line])
    ax.set_xticks([])

plt.ylabel('Singular values of sub-datasets',fontsize=fontsizes[0])
plt.yticks(fontsize=fontsizes[1])
plt.legend(ncol=2,loc='lower left',fontsize=fontsizes[2])

plt.savefig('./Singular_part1.png',dpi=220,bbox_inches='tight')
```

# MW检验

```python
import numpy as np
from scipy.stats import mannwhitneyu
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
MW_S,MW_P = [], []
for k1 in Alls.keys():
    tems,temp = [],[]
    for k2 in Alls.keys():

        # 进行 Mann-Whitney U 检验
        statistic, p_value = mannwhitneyu(Alls[k1], Alls[k2], alternative='two-sided')

        tems.append(statistic)
        temp.append(p_value)
    MW_S.append(tems)
    MW_P.append(temp)

MW_S = np.array(MW_S)
MW_P = np.array(MW_P)
```

```python
colormap = plt.get_cmap('Blues')
colors   = colormap(np.linspace(1, 0.5, Alls['historydata'].shape[0])) # 按百分比

fig = plt.figure(dpi=100,
                    constrained_layout=True,#类似于tight_layout，使得各子图之间的距离自动调整【类似excel中行宽根据内容自适应】
                    figsize=(20,15)
                    )
gs = GridSpec(1, 2, figure=fig, wspace=0.1)#GridSpec将fiure分为3行3列，每行三个axes，gs为一个matplotlib.gridspec.GridSpec对象，可灵活的切片figure


fontsizes = [30, 25, 18]
x = []
for k in Alls.keys():
    x.append(k)

xrange = [i for i in range(len(x))]

ax = fig.add_subplot(gs[0, 0])
im = ax.matshow(MW_S,cmap=colormap)
ax.set_title('Statistic values',fontsize=fontsizes[0])
ax.xaxis.set_ticks_position('bottom')
ax.tick_params(axis='x', labelsize=fontsizes[2])
ax.tick_params(axis='y', labelsize=fontsizes[2])
ax.set_xlabel('Sub-datasets ID, 0 is "Historical data"',fontsize=fontsizes[1])
ax.set_ylabel('Sub-datasets ID, 0 is "Historical data"',fontsize=fontsizes[1])
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size='5%', pad=0.1) 
cbar = fig.colorbar(im, ax=ax, cax=cax)
cbar.ax.tick_params(labelsize=fontsizes[2])

ax = fig.add_subplot(gs[0, 1])
im = ax.matshow(MW_P,cmap=colormap)
ax.set_title('P values',fontsize=fontsizes[0])
ax.xaxis.set_ticks_position('bottom')
ax.tick_params(axis='x', labelsize=fontsizes[2])
ax.tick_params(axis='y', labelsize=fontsizes[2])
ax.set_xlabel('Sub-datasets ID, 0 is "Historical data"',fontsize=fontsizes[1])
ax.set_ylabel('Sub-datasets ID, 0 is "Historical data"',fontsize=fontsizes[1])
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size='5%', pad=0.1) 
cbar = fig.colorbar(im, ax=ax, cax=cax)
cbar.ax.tick_params(labelsize=fontsizes[2])

plt.savefig('./Singular_MWtest.png',dpi=220,bbox_inches='tight')
```
