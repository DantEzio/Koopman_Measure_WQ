```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def smooth(data,avgstep):
    temdata = data.copy()
    for i in range(int(avgstep/2),temdata.shape[0]-int(avgstep/2)):
        temdata[i,:] = np.mean(data[i-int(avgstep/2):i+int(avgstep/2),:],axis=0)
    temdata_se = temdata.copy()
    for i in range(int(avgstep/2),temdata_se.shape[0]-int(avgstep/2)):
        temdata_se[i,:] = np.mean(temdata[i-int(avgstep/2):i+int(avgstep/2),:],axis=0)
    return temdata
```

```python
data = pd.read_csv('./data/data.csv')
sdata = smooth(data.values[:,2:],500)
pd.DataFrame(sdata, columns=data.columns[2:]).to_csv('./data/sdata.csv')
```

```python
k=3
plt.plot(data.values[:,2:][:,k],label='d')
plt.plot(sdata[:,k],label='s')
plt.legend()
```
