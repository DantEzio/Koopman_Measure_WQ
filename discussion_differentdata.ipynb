```python
font1={'family':'Times New Roman','weight':'normal','size':15}
font2={'family':'Times New Roman','weight':'normal','size':10}

dataname = [['Sub dataset 48','Sub dataset 49'],
            ['Sub dataset 55','Sub dataset 56'],
            ['One part of sub dataset 0 (training data)','Sub dataset 66']]
dataid = [['114.csv7','114.csv8'],
          ['75.csv1','75.csv2'],
          ['114.csv0','75.csv9']]

k=-1 #[pH(t)	DO(t)	NH3-N(t)	TOC(t)]
fig,ax=plt.subplots(3,2,figsize=(12,6))
plt.subplots_adjust(hspace=0.5)
for i in range(3):
    for j in range(2):
        ax[i,j].set_title(dataname[i][j],font=font1)
        
        if i == 2:
            ax[i,j].set_xlabel('Time steps',font=font1)
            ax[i,j].plot(data[dataid[i][j]][10:100,k])
        else:
            ax[i,j].plot(data[dataid[i][j]][10:,k])
            
        x1_label = ax[i,j].get_xticklabels() 
        [x1_label_temp.set_fontname('Times New Roman') for x1_label_temp in x1_label]
        y1_label = ax[i,j].get_yticklabels() 
        [y1_label_temp.set_fontname('Times New Roman') for y1_label_temp in y1_label]

plt.savefig('./discussion_differentdata.tif',dpi=200,bbox_inches = 'tight')
```
